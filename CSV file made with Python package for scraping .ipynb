```python
from recipe_scrapers import scrape_me
import csv

header = ['Link', 'Title', 'Total Time', 'Servings', 'Ingredients', 'Instructions']

# open the file in the write mode
with open('C://Users//asus//OneDrive//Documenti//University Docs//MSc Computing//Final Project//RainbowFood(JN)//Rainbow-Food-Collaborative-Filtering-//Recipes.csv', 'w', encoding="utf8", errors = 'ignore') as f:
    # create the csv writer
    writer = csv.writer(f)

    # write the header to the csv file
    writer.writerow(header)
    
    for link in links:
    # iterate over all the links and write down all the info I need    
        scraper = scrape_me(link)
        data = [link, scraper.title(), scraper.total_time(), scraper.yields(), scraper.ingredients(), scraper.instructions()]
        writer.writerow(data)     
```
